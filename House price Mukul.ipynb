# Stratified shuffle 

```python
# The data that is divided may have unequal no of distribution i.e for instance if the values are 0 and1 then
# maybe train data has all 0s and no 1 then 
# while testing the model will fail so we stratify the data
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_set, test_set in split.split(housing, housing['No. of Bedrooms']):
    strat_train_set= housing.loc[train_set]
    strat_test_set= housing.loc[test_set]
```

```python
strat_train_set.info()
```

```python
strat_train_set['No. of Bedrooms'].value_counts()
```
